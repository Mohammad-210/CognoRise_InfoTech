```python
import pandas as pd

# Load mapping data
mapping_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mapping.csv')
```

```python
mapping_df = mapping_df.drop("Unnamed: 0", axis=1)
```

```python
mapping_df.columns = [ 'emoticons','number']
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Train 2.csv')
train_df.head()
```

```python
train_df = train_df.drop("Unnamed: 0", axis=1)
```

```python
train_df.columns = ["TEXT","Label"]
```

```python
train_df
```

```python
import re
import string
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Clean text data
def clean_text(text):
    text = text.lower()
    text = re.sub(f"[{string.punctuation}]", "", text)
    return text

train_df['TEXT'] = train_df['TEXT'].apply(clean_text)

# Tokenize and pad sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_df['TEXT'])
sequences = tokenizer.texts_to_sequences(train_df['TEXT'])
word_index = tokenizer.word_index

max_seq_length = max(len(seq) for seq in sequences)
data = pad_sequences(sequences, maxlen=max_seq_length)

# Encode labels
label_map = {row['number']: i for i, row in mapping_df.iterrows()}
inverse_label_map = {i: row['number'] for i, row in mapping_df.iterrows()}
labels = train_df['Label'].map(label_map).values
```

Split Data


```python
X_train, X_val, y_train, y_val = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

# Define the model
vocab_size = len(word_index) + 1
embedding_dim = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_seq_length),
    LSTM(128, return_sequences=True),
    Dropout(0.5),
    LSTM(128),
    Dropout(0.5),
    Dense(len(label_map), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=32)
```

**Prediction and Evaluation**

```python
# Load test data
test_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Test.csv')
test_df.head()
```

```python
test_df=test_df.drop("Unnamed: 0", axis=1)
```

```python
test_df.head()
```

```python
test_df.columns = ["id",'TEXT']

# Preprocess test data
test_df['TEXT'] = test_df['TEXT'].apply(clean_text)
test_sequences = tokenizer.texts_to_sequences(test_df['TEXT'])
test_data = pad_sequences(test_sequences, maxlen=max_seq_length)

# Make predictions
predictions = model.predict(test_data)
predicted_labels = predictions.argmax(axis=1)
test_df['id'] = [inverse_label_map[label] for label in predicted_labels]

# Save predictions
test_df.to_csv('/content/drive/MyDrive/Colab Notebooks/OutputFormat.csv', index=False)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Predict on validation data for evaluation
val_predictions = model.predict(X_val)
val_predicted_labels = val_predictions.argmax(axis=1)

# Confusion matrix
conf_matrix = confusion_matrix(y_val, val_predicted_labels)

# Classification report
report = classification_report(y_val, val_predicted_labels, target_names=[mapping_df[mapping_df['number'] == id]['emoticons'].values[0] for id in np.unique(y_val)])

print(conf_matrix)
print(report)
```
